```python
import numpy as np
import pandas as pd

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('/Users/kdiepholz/Downloads/Online Retail.xlsx')
```

```python
df.head()
```

```python
df.describe()
```

```python
Basket = (df[df['Country']=="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

Basket.head()
```

```python
def sum_to_boolean(x):
    if x<=0:
        return 0
    else:
        return 1

Basket_Final = Basket.applymap(sum_to_boolean)
```

```python
Basket_Final.drop('POSTAGE', inplace=True, axis=1)

Basket_Final.head()
```

```python
Frequent_itemsets = apriori(Basket_Final, min_support = 0.06, use_colnames = True)

Frequent_itemsets.sort_values('support', ascending = False).head()
```

Using the apriori algorithm, we can see that the (RABBIT NIGHT LIGHT) has the highest support, with .160521

```python
Frequent_itemsets = apriori(Basket_Final, min_support = 0.05, use_colnames = True)

Frequent_itemsets.sort_values('support', ascending = False).head()
```

We set the minimum support threshhold to 5%, or .05 as min_support

Using this, we can run the association_rules module with our updated frequent itemsets

```python
Asso_Rules = association_rules(Frequent_itemsets, metric = "lift", min_threshold = 8)
Asso_Rules.sort_values('lift',ascending = False)
```

From this we can see that **Childrens cutlery spaceboy and childrens cutlery dolly girl** are purchased together. 
We can also see that **alarm clock bakelike green and alarm clock bakelike red** are purchased together. 
Finally we can see that **Set/6 red spotty paper cups and set/6 red spotty paper plates** are purchased together. 

```python
Asso_Rules = association_rules(Frequent_itemsets, metric = "confidence", min_threshold = .8)
Asso_Rules.sort_values('confidence',ascending = False)
```

From our confidence association rules, we can see that **set/6 red spotty paper cups and set/6 red spotty paper plates** and **set/20 red retrospot paper napkins and set/6 red spotty paper cups** and **childrens cutlery spaceboy and childrens cutlery dolly girl** are all purchased together.

The rule with the highest lift is not the same as the rule with the highest confidence. This could be due to a number of reasons. One of the largest reasons is because lift and confidence take on a different equation. In addition, they take into account different things, with lift being the highest number of times something was purchased together and confidence being the ratio between how many times it was purchased together vs the number of times it was purchased in general.
